In this file we will draw some basic visuals about our dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

loading dataset and a preview of dataset:-

```python
df = pd.read_csv("global_movies_dataset_1950_2026.csv")
print(df.head())
```

We will create some basic visuals to showcase trends in our dataset
......lets begin with a bar chart of categories(genres) showing the top ten movie genres by count.

```python
df['genre'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Movie Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

Now lets make a line chart about the number of movies released each year

```python
movies_per_year = df['release_year'].value_counts().sort_index()

plt.plot(movies_per_year)
plt.title("Movies Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()
```

Another interesting visual will be count of movies of a particular rating(audience score) and a graph(histogram) depecting that for lets say 10 possible buckets.

```python
plt.hist(df['audience_score'].dropna(), bins=10)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

This concludes some of basic visuals, now lets move on to better visuals using seaborn
first one will be a boxplot for genre vs ratings which will show us the outliers.

```python
sns.boxplot(x='genre', y='audience_score', data=df)
plt.xticks(rotation=45)
plt.show()
```

Another one is a heatmap of the dataset as shown below.

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
