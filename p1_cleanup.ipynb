We will start with cleanup of the raw data which includes following steps.

Step 1: File is to be loaded for begining of the cleanup process.

```python
import pandas as pd

df = pd.read_csv("movies.csv")

# to confirm loading and for Preview of the dataset
print(df.head())
print(df.info())
```

step 2:check for null values and then replace the null values with median or "unknown" depending upon the type of data.

```python
print(df.isnull().sum())
```

we can clearly see that the null values are present in rating, box office and director.
rating and box office are of numerical type thus can be replaced by median of the column while as director being a text type can be replaced by "unknown".

```python
# Fill numeric columns with median
df['Rating']=df['Rating'].fillna(df['Rating'].median())
df['BoxOffice']=df['BoxOffice'].fillna(df['BoxOffice'].median())

# Fill categorical column with "Unknown"
df['Director']=df['Director'].fillna("Unknown")
print(df.isnull().sum())
```

We can clearly see that null values have been removed successfully.

step 3: now the next problem is to remove duplicate values.

```python
# Find duplicates based on a specific column
duplicates = df[df.duplicated(subset=['Title'])]
print(duplicates)

# drop duplicates based on that specific column
df=df.drop_duplicates(subset=['Title'])

# recheck the duplicate values(should be empty)
duplicates = df[df.duplicated(subset=['Title'])]
print(duplicates)
```

Step 4: Next step is to fix the data type if necessary
.......release year from int to date

```python
df['ReleaseYear'] = pd.to_datetime(df['ReleaseYear'], format='%Y')
```

```python
print(df.dtypes)
```

All the steps are done, now we need to save the dataframe to a csv file

```python
df.to_csv("cleaned_movies_dataset.csv", index=False)
```
